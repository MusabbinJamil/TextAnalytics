```python
# 1.8B Q4 Qwen
model1_name = "Qwen/Qwen1.5-0.5B"  #Qwen-1.8B
tokenizer = AutoTokenizer.from_pretrained(model1_name)
model = AutoModelForCausalLM.from_pretrained(model1_name, torch_dtype="auto", device_map="auto")
```

```python
email = """
Dear WidgetCorp,
I am writing to express my deep disappointment with your service regarding my recent order (Order
#98765). I had ordered the "SuperWidget" as a birthday gift for my daughter, Emily, whose birthday
was on March 15th. Despite your website promising delivery within three days, the product arrived
only yesterday, missing her birthday entirely.
I attempted to contact your customer service multiple times for an explanation and to expedite the
delivery, but to no avail. My calls and emails went unanswered, leaving me in a difficult position with
no gift for Emily’s special day.
This experience has left a sour taste, and I am reconsidering future purchases from WidgetCorp. I
expect a prompt explanation and an assurance that such delays will not occur in the future.
Sincerely,
Jane Doe
"""
```

```python
summary_prompt = f"Summarize the following customer email briefly:\n\n{email}\n\nSummary:"
inputs = tokenizer(summary_prompt, return_tensors="pt", truncation=True)
outputs = model.generate(inputs["input_ids"], max_length=150)
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Summary:\n", summary)
```

```python
qa_prompt = f"""
Customer Email: {email}
For question-answering tasks, use the following two questions:
1. What was the reason for the customer's disappointment with WidgetCorp?
2. How did WidgetCorp's customer service respond to the customer's inquiries about the late delivery?
Answer:
"""
inputs = tokenizer(qa_prompt, return_tensors="pt", truncation=True)
outputs = model.generate(inputs["input_ids"], max_length=200)
qa_response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("QA Response:\n", qa_response)
```

```python
ner_prompt = f"Identify the named entities in the following email (such as names, dates, products, order numbers):\n\n{email}\n\nEntities:"
inputs = tokenizer(ner_prompt, return_tensors="pt", truncation=True)
outputs = model.generate(inputs["input_ids"], max_length=150)
ner_response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Named Entities:\n", ner_response)
# no result after running from for 26 minutes
```

```python
# English to French
translation_prompt = f"Translate the following email to French:\n\n{email}\n\nFrench Translation:"
inputs = tokenizer(translation_prompt, return_tensors="pt", truncation=True)
outputs = model.generate(inputs["input_ids"], max_length=300)
french_translation = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("French Translation:\n", french_translation)

# French back to English
back_translation_prompt = f"Translate the following back to English:\n\n{french_translation}\n\nEnglish Translation:"
inputs = tokenizer(back_translation_prompt, return_tensors="pt", truncation=True)
outputs = model.generate(inputs["input_ids"], max_length=300)
back_to_english = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Back to English:\n", back_to_english)
```
